# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Getting the Data

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

# Setting up the DataFrame

```python
df_feat = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df_feat.head()
```

```python
cancer['target'] 
```

```python
cancer['target_names']
```

```python
df_feat['target']= cancer.target
df_feat.head(25)
```

```python
df_feat[df_feat['target']==0].head() 
```

```python
cancer['target_names']
```

```python
df_feat['result_name']= df_feat['target'].apply(lambda x: cancer['target_names'][x]) 
```

```python
df_feat.head(25)
```

# Exploratory Data Analysis

```python
df0= df_feat[df_feat['target']==0]
df1= df_feat[df_feat['target']==1] 
```

```python
df0.head() 
```

```python
df1.head()
```

```python
plt.xlabel('mean radius')
plt.ylabel('mean texture') 
plt.scatter(df0['mean radius'], df0['mean texture'], color= 'green', marker= '+')
plt.scatter(df1['mean radius'], df1['mean texture'], color= 'red', marker= '+') 
```

```python
plt.xlabel('mean perimeter')
plt.ylabel('mean area') 
plt.scatter(df0['mean perimeter'], df0['mean area'], color= 'green', marker= '+')
plt.scatter(df1['mean perimeter'], df1['mean area'], color= 'red', marker= '+') 
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X= df_feat.drop(['target','result_name'], axis='columns')
y= df_feat['target'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

# Train the Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
model = SVC(gamma=0.0001)
```

```python
model.fit(X_train,y_train) 
```

# Predictions

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```
